```python
# seperate words with mutliple tags
sentence_tag_dict = {}
tags_list = []
tag_pair_dict = {}
unique_tags = {"STRT":0}
word_tag_pair = {}
with open("POS.train", "r") as text:
    for line in text:
        parts = line.split()
        sentence = []
        tags = []
        tag1 = "STRT"
        unique_tags[tag1] += 1
        for element in parts:
            items_list = element.rsplit('/',1)
            word = items_list[0]
            tag2 = items_list[1]
            sentence.append(word)
            tags.append(tag2)
            # creating tag_pair
            if tag2 not in tag_pair_dict.keys():
                tag_pair_dict[tag2] = {tag1: [1]}
            else:
                if tag1 not in tag_pair_dict[tag2].keys():
                    tag_pair_dict[tag2][tag1] = [1]
                else:
                    tag_pair_dict[tag2][tag1][0] += 1
            
            
            # finding tag_counts
            if tag2 not in unique_tags.keys():
                unique_tags[tag2] = 1
            else:
                unique_tags[tag2] += 1
            
            # finding word_tag_pair
            if word not in word_tag_pair.keys():
                word_tag_pair[word] = {tag2: [1]}
            else:
                if tag2 in word_tag_pair[word].keys():
                    word_tag_pair[word][tag2][0] += 1
                else:
                    word_tag_pair[word][tag2] = [1]
            
            tag1 = tag2
        sentence_tag_dict[tuple(sentence)] = tags

# word_tag_pair
# unique_tags
# tag_pair_dict
# sentence_tag_dict
```

```python
# finding tag_pair_prob

tag_pair_dict
for tag2 in tag_pair_dict.keys():
    for tag1 in tag_pair_dict[tag2].keys():
        prob = tag_pair_dict[tag2][tag1][0] / unique_tags[tag1]
        tag_pair_dict[tag2][tag1].append(prob)
# tag_pair_dict
```

```python
# finding word_tag_prob
# word_tag_pair
for word in word_tag_pair.keys():
    for tag in word_tag_pair[word].keys():
        prob = word_tag_pair[word][tag][0] / unique_tags[tag]
        word_tag_pair[word][tag].append(prob)
# word_tag_pair
```

```python
sentences = list(sentence_tag_dict.keys())
# tag_pair_dict['NP']['STRT'][1]
```

```python
def max_calculate(score_list, prev_word_idx, curr_tag_idx, tag2):
    max_list = []
    max_num = 0
    max_idx = 0
    for t_idx, tag in enumerate(tag_list):
        if (t_idx, prev_word_idx) in score_list.keys():
            if tag2 in tag_pair_dict.keys():
                if tag1 in tag_pair_dict[tag2].keys():
                    if max_num < (score_list[(t_idx, prev_word_idx)] * tag_pair_dict[tag2][tag1][1]):
                        max_num = score_list[(t_idx, prev_word_idx)] * tag_pair_dict[tag2][tag1][1]
                        max_idx = t_idx
        
        
    return max_idx, max_num
```

```python
# Veterbi Algorithm
back_ptr = {}
for number, sent in enumerate(sentences):
    
    tag_list = list(unique_tags.keys())
    score_prob = {}
    # Initialization Step
    for index, tag in enumerate(tag_list):
        if tag in word_tag_pair[sent[0]].keys():
            if 'STRT' in tag_pair_dict[tag].keys():
                score_prob[(index, 1)] = word_tag_pair[sent[0]][tag][1] * (tag_pair_dict[tag]['STRT'][1])
                back_ptr[(index,1)] = 0
    
    #Iteration step
    for w_idx, word in enumerate(sent[1:]):
        for t_idx, tag in enumerate(tag_list):
            max_idx, max_score = max_calculate(score_prob, w_idx-1, t_idx, tag)
            if tag in word_tag_pair[sent[w_idx]].keys():
                score_prob[(t_idx, w_idx)] = word_tag_pair[sent[w_idx]][tag][1] * max_score
                back_ptr[(t_idx, w_idx)] = max_idx
            
```
